```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import torch
import torch.nn as nn
from mmdet3d.apis import init_model, inference_detector
import struct
from natsort import natsorted
from glob import glob
from open3d import *
import pathlib
```

```python
config_file = '/content/mmdetection3d/configs/pointpillars/pointpillars_hv_secfpn_8xb6-160e_kitti-3d-3class.py' # Path to the config file

checkpoint_file = '/content/drive/MyDrive/188/hv_pointpillars_secfpn_6x8_160e_kitti-3d-3class_20220301_150306-37dc2420.pth' # Path to pretrained checkpoint

# Build the model from a config file and load checkpoint
model = init_model(config_file, checkpoint_file, device='cuda:0')
```

```python
print(model)
```

Added a custom hook to save the 2D pseudo-image created by PointPillars - had to split into smaller sets of 500 due to memory issues

```python
point_cloud_files = natsorted(glob("/content/drive/MyDrive/188/kitti/training/velodyne_reduced/*"))

encoder_layer = model.middle_encoder

feature_maps = []
def hook(module, input, output):
  feature_maps.append(output.squeeze(0))

handle = encoder_layer.register_forward_hook(hook)

with torch.no_grad():
  for point_cloud in point_cloud_files[1000:]:
    output = inference_detector(model, point_cloud)

handle.remove()
```

```python
print(len(feature_maps))
```

```python
print(torch.count_nonzero(feature_maps[400]))
```

```python
for index, point_cloud in enumerate(point_cloud_files[1000:]):
  path = pathlib.Path(point_cloud)
  index = path.parts.index('kitti')
  new_path = pathlib.Path(*path.parts[:index]).joinpath('train_embeds').joinpath(*path.parts[-1:]).with_suffix('.pt')

  torch.save(feature_maps[index], new_path)
```
